```python
# filtering data frame by cities with the "Cfa" climate classifcation, and population similar to Melbourne
pop_max= 5000000
pop_min= 4000000

classification = "Cfa"

filter_df = df.loc[(df['Climate Classification']==classification) & (df['Population'] >=pop_min) & (df['Population'] < pop_max)]
filter_df
```
